**1.Demonstrate the use of inheritance.**


```python
class Person(object):
	def __init__(self, name):
		self.name = name
	def getName(self):
		return self.name

	def isEmployee(self):
		return False

class Employee(Person):

	def isEmployee(self):
		return True


emp = Person("Shruti") 
print(emp.getName(), emp.isEmployee())

emp = Employee("Sneha") 
print(emp.getName(), emp.isEmployee())
```

**2.Multilevel Inheritance.**

```python
class Base(object):
	def __init__(self, name):
		self.name = name
	def getName(self):
		return self.name

class Child(Base):
	def __init__(self, name, age):
		Base.__init__(self, name)
		self.age = age
	def getAge(self):
		return self.age

class GrandChild(Child):
	
	def __init__(self, name, age, address):
		Child.__init__(self, name, age)
		self.address = address
	def getAddress(self):
		return self.address	

g = GrandChild("Shruti", 23, "Latur")
print(g.getName(), g.getAge(), g.getAddress())
```

**3.Multiple Inheritance.**

```python
class Class1:
    def m(self):
        print("In Class1")
       
class Class2(Class1):
    def m(self):
        print("In Class2")
 
class Class3(Class1):
    def m(self):
        print("In Class3") 
        
class Class4(Class2, Class3):
    pass  
     
obj = Class4()
obj.m()
```
